```python
import pandas as pd
import table_cleaning_functions as tcf
```

```python
df = pd.read_csv("../home/data/raw_data/MW/MW.tsv", sep="\t")
df.head()
```

```python
df["evidence_source"] = df["evidence_source"].apply(lambda x: "Pubmed:" + x if ":" not in x else x)
df.head()
```

```python
df = df.rename(columns={"assessed_biomarker_entity_Id": "assessed_biomarker_entity_id"})
df.columns
```

```python
for col in df.columns:
    if col in {"assessed_entity_type", "best_biomarker_role", "specimen"}:
        df[col] = tcf.make_lowercase(df[col])
df.head()
```

```python
df.to_csv("../home/data/cleaned_data/MW/mw.tsv", sep="\t", index=False)
```
