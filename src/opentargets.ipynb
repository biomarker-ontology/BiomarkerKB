```python
import pandas as pd
import table_cleaning_functions as tcf
```

```python
df = pd.read_csv('../home/data/processed_data/GlyGen/opentargets/opentargets_processed.tsv', sep = '\t')
df.head()
```

```python
df['biomarker_id'] = df.groupby(['biomarker', 'assessed_biomarker_entity_id', 'assessed_biomarker_entity', 'condition']).ngroup()
df.head()
```

```python
print(df.shape)
df['biomarker_id'].value_counts()
```

```python
df = df.drop_duplicates()
print(df.shape)
df['biomarker_id'].value_counts()
```

```python
df.head()
```

```python
for col in df.columns:
    df[col] = tcf.strip_values(df[col])
df.head()
```

```python
df['tag'] = tcf.strip_internal_whitespace(df['tag'])
df['tag'] = tcf.make_lowercase(df['tag'])
df.head()
```

```python
df.to_csv('../home/data/cleaned_data/GlyGen/opentargets.tsv', sep = '\t', index = False)
```
